```python
# df['n_hos_beds'].fillna(df['n_hos_beds'].mean(), inplace=True)
df = df.assign(n_hos_beds=df['n_hos_beds'].fillna(df['n_hos_beds'].mean()))
```

```python
# df['waterbody'].fillna(df['waterbody'].mode()[0], inplace=True)
df = df.assign(waterbody=df['waterbody'].fillna(df['waterbody'].mode()[0]))
```
