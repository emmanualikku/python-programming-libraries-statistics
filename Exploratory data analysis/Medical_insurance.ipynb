```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["age"],bins=20,kde=True)
plt.title("Medical_insurance.csv")
plt.show()     
print(df['age'].describe())
sns.kdeplot(df["age"], shade=True)
plt.title("KDE Plot of age")
plt.show()
```

```python
sns.histplot(df["bmi"],bins=20,kde=True)
plt.title("Medical_insurance.csv")
plt.show()     
print(df['bmi'].describe())
sns.kdeplot(df["bmi"], shade=True)
plt.title("KDE Plot of bmi")
plt.show()
```

```python
sns.histplot(df["children"],bins=20,kde=True)
plt.title("Medical_insurance.csv")
plt.show()     
print(df['children'].describe())
sns.kdeplot(df["children"], shade=True)
plt.title("KDE Plot of children")
plt.show()
```

```python
sns.histplot(df["charges"],bins=20,kde=True)
plt.title("Medical_insurance.csv")
plt.show()     
print(df['charges'].describe())
sns.kdeplot(df["charges"], shade=True)
plt.title("KDE Plot of charges")
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x=df["sex"])
plt.xticks(rotation=40)
plt.title("sex Count")
plt.show()
print(df['sex'].describe())
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x=df["smoker"])
plt.xticks(rotation=40)
plt.title("smoker Count")
plt.show()
print(df['smoker'].describe())
```
