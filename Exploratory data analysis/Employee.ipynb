```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(df["JoiningYear"],bins=20,kde=True)
plt.title("Employee.csv")
plt.show()     
print(df['JoiningYear'].describe())
sns.kdeplot(df["JoiningYear"], shade=True)
plt.title("KDE Plot of JoiningYear")
plt.show()
```
